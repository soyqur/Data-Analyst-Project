```python
import pandas as pd
import os
```

```python
BASE_DIR = os.path.dirname(os.getcwd())

df = pd.read_csv(os.path.join(BASE_DIR, "data", "raw", "europe_sales.csv"))

df
```

```python
df.info()
```

```python
df.describe()
df.isnull().sum()
df.shape
```

```python
date_cols = ['Order Date', 'Date', 'Ship Date']

for col in date_cols:
    df[col] = pd.to_datetime(df[col], format='mixed')
```

```python
string_cols = df.select_dtypes(include='object').columns

for col in string_cols:
    df[col] = (
        df[col]
        .str.replace(r'[^a-zA-Z\s]', '', regex=True)  # remove symbols + numbers
        .str.replace(r'\s+', ' ', regex=True)         # fix spaces
        .str.strip()
    )
```

```python
file_path = os.path.join(BASE_DIR, "data", "processed", "cleaned_europe_sales.csv")

df.to_csv(file_path,index=False)
```
